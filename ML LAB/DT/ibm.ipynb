```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler 
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('ibm.csv')
```

```python
df.head()
```

```python
df.head()
```

```python
def dummyEncode(df):
    column = list(df.select_dtypes(include=['category','object']))
    for x in column:
        dummies = pd.get_dummies(df[x],prefix = x,dummy_na = False)
        df = df.drop(x,1)
        df = pd.concat([df,dummies],axis=1)
    return df
```

```python
x = df.drop('Attrition',axis=1)
y = df['Attrition']
x = dummyEncode(x)
y = y.replace({"Yes":1,"No":0})
```

```python
x.head()
```

```python
import sklearn.feature_selection
select = sklearn.feature_selection.SelectKBest(k=20)
```

```python
x,xt,y,yt = train_test_split(x,y,test_size=.1,random_state=1)
```

```python
col = select.fit(x,y)

index = col.get_support(indices=True)
cols = [x.columns[i] for i in index]
```

```python
x = x[cols]
xt = xt[cols]
```

```python
from imblearn.over_sampling import SMOTE
```

```python
oversampler=SMOTE(random_state=0)
x,y = oversampler.fit_sample(x,y)
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
from sklearn.model_selection import cross_val_score,cross_val_predict
```

```python
def print_score(clf, X_train, y_train, X_test, y_test, train=True):
    '''
    print the accuracy score, classification report and confusion matrix of classifier
    '''
    if train:
        '''
        training performance
        '''
        print("Train Result:\n")
        print("accuracy score: {0:.4f}\n".format(accuracy_score(y_train, clf.predict(X_train))))
        print("Classification Report: \n {}\n".format(classification_report(y_train, clf.predict(X_train))))
        print("Confusion Matrix: \n {}\n".format(confusion_matrix(y_train, clf.predict(X_train))))

        res = cross_val_score(clf, X_train, y_train, cv=10, scoring='accuracy')
        print("Average Accuracy: \t {0:.4f}".format(np.mean(res)))
        print("Accuracy SD: \t\t {0:.4f}".format(np.std(res)))
        
    elif train==False:
        '''
        test performance
        '''
        print("Test Result:\n")        
        print("accuracy score: {0:.4f}\n".format(accuracy_score(y_test, clf.predict(X_test))))
        print("Classification Report: \n {}\n".format(classification_report(y_test, clf.predict(X_test))))
        print("Confusion Matrix: \n {}\n".format(confusion_matrix(y_test, clf.predict(X_test))))  
```

```python
clf = DecisionTreeClassifier(random_state=42)
clf.fit(x,y)
```

```python
print_score(clf,x,y,xt,yt,train=True)
```

```python
print_score(clf,x,y,xt,yt,train=False)
```
